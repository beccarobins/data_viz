```python
import networkx as nx
import matplotlib.pyplot as plt

# 1. Create a directed graph for the WBS
wbs = nx.DiGraph()

# 2. Add Nodes (Tasks)
nodes = ["Project", "Plan", "Design", "Code", "Task 1.1", "Task 1.2", "Task 2.1"]
wbs.add_nodes_from(nodes)

# 3. Add Edges (Hierarchical relationships: Parent -> Child)
edges = [
    ("Project", "Plan"),
    ("Project", "Design"),
    ("Project", "Code"),
    ("Plan", "Task 1.1"),
    ("Plan", "Task 1.2"),
    ("Design", "Task 2.1")
]
wbs.add_edges_from(edges)

# 4. Define a hierarchical layout (using custom positioning for a tree)
# For better tree layouts, consider: pos = nx.nx_agraph.graphviz_layout(wbs, prog='dot')
pos = {
    "Project": (0, 2.5),
    "Plan": (-1, 2),
    "Design": (0, 2),
    "Code": (1, 2),
    "Task 1.1": (-1.5, 1.5),
    "Task 1.2": (-0.5, 1.5),
    "Task 2.1": (0, 1.5)
}

# 5. Draw the graph
plt.figure(figsize=(10, 6))
nx.draw(wbs, pos, with_labels=True, node_size=2000, node_color="skyblue", 
        font_size=10, font_weight="bold", arrowsize=20)
plt.title("Work Breakdown Structure")
plt.show()
```
